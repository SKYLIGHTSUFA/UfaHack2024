```python
import os
from imutils import paths
import cv2
```

```python
def variance_of_laplacian(image):
	return cv2.Laplacian(image, cv2.CV_64F).var()
```

```python
dir_ = 'C://Users//fatik//PycharmProjects//UfaHack2024//data//bloggers'
threshold = 100
i = 0
for imagePath in paths.os.listdir(dir_):
    for image_name in os.listdir(os.path.join(dir_, imagePath )):
        image = cv2.imread(os.path.join(dir_, imagePath, image_name))
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        fm = variance_of_laplacian(gray)
        text = "Not Blurry"
        if fm < threshold:
            i+=1
            text = "Blurry"
            os.remove(os.path.join(dir_, imagePath, image_name))
            print(text, i)
        cv2.putText(image, "{}: {:.2f}".format(text, fm), (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 3)
print(i)
```
