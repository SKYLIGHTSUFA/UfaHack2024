```python
import cv2
from deepface import DeepFace
import os
from mtcnn import MTCNN
```

```python
models = [
  "VGG-Face",
  "Facenet",
  "Facenet512",
  "OpenFace",
  "DeepFace",
  "DeepID",
  "ArcFace",
  "Dlib",
  "SFace",
  "GhostFaceNet",
]
backends = [
  'opencv',
  'ssd',
  'dlib',
  'mtcnn',
  'retinaface',
  'mediapipe',
  'yolov8',
  'yunet',
  'fastmtcnn',
]
```

```python
from tqdm.notebook import tqdm
detector = MTCNN()
dicter3 = {}
embeddings1 = []
z = 0
dir_ = r'C:\Users\fatik\PycharmProjects\UfaHack2024\data\bloggers'
for filename in tqdm(os.listdir(dir_)):
  z+=1
  print(z)
  if z==500:
    break
  i = 0
  for image in os.listdir(os.path.join(dir_, filename)):
    i+=1
    img = cv2.cvtColor(cv2.imread(os.path.join(dir_, filename, image)), cv2.COLOR_BGR2RGB)
    detections = detector.detect_faces(img)
    if len(detections) > 1:
      continue
    for detection in detections:
      confidence = detection["confidence"]
      if confidence > 0.9:
        x, y, w, h = detection["box"]
        detected_face = img[int(y):int(y + h), int(x):int(x + w)]
        embedding = DeepFace.represent(detected_face, model_name='Facenet', enforce_detection=False)
        dicter3[f'{filename}_{i}'] = embedding[0]["embedding"]
```

```python
import pickle
with open('actors_ussr_embeddings.pkl', 'wb') as f:
    pickle.dump(dicter3, f)
```
